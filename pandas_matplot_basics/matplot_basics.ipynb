```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[7,4,5,6]
plt.plot(x,y)
plt.xlabel('yoy')
plt.ylabel('toyo')
plt.show()
```

```python
x=[1,2,3,4]
y=[7,4,5,6]
plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel('yoy')
plt.ylabel('toyo')
plt.legend(["This is A","This is B"])
plt.show()
```

```python
x=[1,2,3,4]
y=[7,4,5,6]
plt.plot(y,x)
plt.plot(x,y)
plt.xlabel('yoy')
plt.ylabel('toyo')
plt.legend(["This is A","This is B"])
plt.show()
```

```python
x=[1,2,3,4]
y=[7,4,5,6]
plt.bar(x,y)
plt.bar(y,x)
plt.xlabel('yoy')
plt.ylabel('toyo')
plt.legend(["This is A","This is B"])
plt.show()
```

```python
y=[34,44,3,4]
plt.hist(y)
plt.xlabel('yoy')
plt.ylabel('toyo')
plt.legend(["This is A","This is B"])
plt.show()
```

```python
x=[7,4,5,6]
label=['1st','2nd','3rd','4th']
plt.pie(x,labels=label)
plt.show()
```

```python
#ploting with formating

month=['jan','feb','march','april']
India=[1000,2333,4444,4444444]
BD=[2333,566,43333,544044]
plt.plot(month,India,color='orange',marker='*',markersize=12,label='India')
plt.plot(month,BD,color='green',linestyle='dashed',markersize=12,label='Bangladesh')
plt.legend()
plt.show()
```
